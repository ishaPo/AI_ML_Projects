```python
from transformers import pipeline # for pipeline creation
import pandas as pd # for data handling
```

```python
pline = pipeline(task = 'table-question-answering',model ='google/tapas-base-finetuned-wtq')
```

```python
# importing the csv data
# Dataset -> Sustainable Development Data up to 2023
# Link to Dataset -> https://www.kaggle.com/datasets/sazidthe1/sustainable-development-report
path = '/content/sustainable_development_report_2023.csv'

table = pd.read_csv(path)
```

```python
table
```

```python
table.describe()
```

```python
# taking any 10 rows at random as model takes limited data
# Use the sample() method to select 10 random rows
random_10_rows = table.sample(n=10, random_state=42)  # Set random_state for reproducibility
random_10_rows.reset_index(drop=True, inplace=True) # reset indexing
random_10_rows
```

```python
random_10_rows = random_10_rows.astype(str)
```

```python
random_10_rows
```

# Question and Answering

* As of now the Model can answer basic questions on limited data
* For it to be able to answer questions on more data, we can use a model capabale of taking more data as input, or we can perform the operation in batches in future versions

```python
query = "Which country has highest Overall Score?" # we will ask it a question
```

```python
ans = pline(table = random_10_rows, query = query)['answer']
print(ans)
```
